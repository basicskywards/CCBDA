```python
import csv
import time
import collections
```

```python
def read_db(file_path):
    db = []
    with open(file_path, newline = '') as csvfile:
        rows = csv.reader(csvfile)
        for row in rows:
            db.append(row)
        return db

class prefixspan(object):
    
    def __init__(self, database, min_sup = 2):
        self.final_sequences = []
        self.database = database
        self.min_sup = min_sup
    
    def get_prefix(self, items, row_pos_list):
        # create prefix & index of suffix
        
        support = len(row_pos_list)
        item_support = (items, support)
        self.final_sequences.append(item_support)
        frequent_placeholder = collections.defaultdict(list)
        
        for idx, row_pos in enumerate(row_pos_list):
            row = row_pos[0]
            pointer_position = row_pos[1]
            tmp_sequence = self.database[row]
            
            for item_position in range(pointer_position + 1, len(tmp_sequence)):
                # suffix potision loop
                event = tmp_sequence[item_position]
                
                #record_row_position = frequent_placeholder[event]
                record_row_position = frequent_placeholder[event]
                
                support = len(record_row_position)
                
                #row_in_previous_loop = record_row_position[-1][0]
                
                if support == 0 or record_row_position[-1][0] != row:
                    # check frequency of sequence
                    #record_row_position = frequent_placeholder[event]
                    record_row_position.append((row, item_position))
                    
        return frequent_placeholder  
    
    def run_prefixspan(self, items, row_pos_list):
        # main prefixspan
        frequent_placeholder_temp = self.get_prefix(items, row_pos_list)
        
        for item, row_item_position in frequent_placeholder_temp.items():
            support_count = len(row_item_position)
            
            # check sequence frequent or not
            if self.min_sup <= support_count: 
                # recursive to find k+1 frequent sequences 
                self.run_prefixspan(items + [item], row_item_position)

def output_txt(file_name, final_freq_seqs):
    with open(file_name, "w") as f:
        for items, support in final_freq_seqs:

            tmp = len(items)
            f.write("[")
            for idx, item in enumerate(items):

                if tmp == 1:
                    f.write("%s],%s\n"%(item, str(support)))

                else:

                    if idx != (tmp-1):
                        f.write("%s,"%(item))

                    elif idx == (tmp-1):
                        f.write("%s],%s\n"%(item, str(support)))
```

```python
tStart = time.time()
# dataset_1
# input file: dataset_1.csv
# output_file: output_1.txt
support = 2
file_input = 'publicdataset_2.csv'
file_output = 'public_output_2.txt'

# read input
db = read_db(file_input)

# execute algorithm
results = []
tmp = [(i, -1) for i in range(len(db))]
min_freq_seq = prefixspan(db, support)
min_freq_seq.run_prefixspan([], tmp)
final_freq_seq = min_freq_seq.final_sequences

# write output
output_txt(file_output, sorted(final_freq_seq[1:]))

tEnd = time.time()
print('\nRun time: {}'.format((tEnd - tStart)))
```
